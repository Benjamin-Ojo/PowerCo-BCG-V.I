```python
def plot_stacked_bars(dataframe, title_, size_=(18, 10), rot_=0, legend_="upper right"):
    """
    Plot stacked bars with annotations
    """
    ax = dataframe.plot(
        kind="bar",
        stacked=True,
        figsize=size_,
        rot=rot_,
        title=title_
    )

    # Annotate bars
    annotate_stacked_bars(ax, textsize=14)
    # Rename legend
    plt.legend(["Retention", "Churn"], loc=legend_)
    # Labels
    plt.ylabel("Company base (%)")
    plt.show()

def annotate_stacked_bars(ax, pad=0.99, colour="white", textsize=13):
    """
    Add value annotations to the bars
    """

    # Iterate over the plotted rectanges/bars
    for p in ax.patches:
        
        # Calculate annotation
        value = str(round(p.get_height(),1))
        # If value is 0 do not annotate
        if value == '0.0':
            continue
        ax.annotate(
            value,
            ((p.get_x()+ p.get_width()/2)*pad-0.05, (p.get_y()+p.get_height()/2)*pad),
            color=colour,
            size=textsize
        )
```

```python
# Checking the Normalized Distribution 

def plot_distribution(dataframe, column, ax, bins_=50):
    """
    Plot variable distirbution in a stacked histogram of churned or retained company
    """
    # Create a temporal dataframe with the data to be plot
    temp = pd.DataFrame({"Active": dataframe[dataframe["churn"]==0][column],
    "Churn":dataframe[dataframe["churn"]==1][column]})
    # Plot the histogram
    temp[["Active","Churn"]].plot(kind='hist', bins=bins_, ax=ax, stacked=True)
    # X-axis label
    ax.set_xlabel(column)
    # Change the x-axis to plain style
    ax.ticklabel_format(style='plain', axis='x')
```

```python
fig, axs = plt.subplots(nrows=4, figsize=(18, 25))

plot_distribution(client_df, 'cons_12m', axs[0])
plot_distribution(client_df[client_df['has_gas'] == 't'], 'cons_gas_12m', axs[1])
plot_distribution(client_df, 'cons_last_month', axs[2])
plot_distribution(client_df, 'imp_cons', axs[3])
```

## Price Sencitivity Hypothesis Testing
---

In this section we will be defining what price sencitivity, how we are mearsuring it with our dataset, and explaining how changes in price of our services can affect the churning rate of our client. 

### What is Price Sensitivity?

Price sensitivity can be define as the degree of responses of a feature to a change in the price. 

In testing our client hypothesis on whether churn has a significant relationship with change in price, we will be deriving the correlation between price and churn over time. 

### Measuring rice Sensitivity to Churn Rate.

To measure the price sensitivity of our data we will be measuring correlative relationship that exist between churn and price. Base off the result we get from this measurement we will be able to determine whether there exist a negative or positive relationship between churn and price. If the relationship between the churn and price is positive, we will not reject the null hypothesis stating that there exist a relationship between churn and price. Other wise, we will accept the Alternative Hypothesis that there exist no relatonship between churn and price. The following steps will be taken to clean, sort, and measure our data in other to test our client hypothesis:



*   STEP 1: Convert the `price_date` column into date_time formate. This is will make it easy to sort and manupulate our data according to the date of each customer.

*   STEP 2: Get the average price for a year, half the year, and three month( a quarter of a year).

*   STEP 3: Change the names of all columns in the price dataset to reflect the new dataset, after this we will add the price dataset `price_off_peak_ver` with `price_off_peak_fix` and do the same for the ramaining 4 columns. 

*   STEP 4: Marge the a`client_churn` and `new_price`dataset into one dataframe using the `i.d` as the primary key. 

*   STEP 5: Perform a correlation on the new marge data to measure the relationship between churn and price. Following this we will create a heat map to visualize this relationship.

*   STEP 6: Marge the `price_analysis` dataset with the `client_df` and dropping the churn column.

```python
"""
STEP 1: CONVERTING THE STRICK DATASET TO A DATE_TIME FORMATE.
"""
client_df['date_activ'] = pd.to_datetime(client_df['date_activ'])
client_df['date_end'] = pd.to_datetime(client_df['date_end'])
client_df['date_modif_prod'] = pd.to_datetime(client_df['date_modif_prod'])
client_df['date_renewal'] = pd.to_datetime(client_df['date_renewal'])
price_df['price_date'] = pd.to_datetime(price_df['price_date'])

client_df.info()
price_df.info()

```

```python
"""
STEP 2: GETTING AVERAGES PRICE FOR EACH MONTH.
"""
# Calculating the Average Price for the Verious Time Span. 
average_year_price = price_df.groupby('id').mean().reset_index()
average_6m_price = price_df[price_df['price_date'] >= '2015-07-01'].groupby('id').mean().reset_index()
average_3m_price = price_df[price_df['price_date'] >= '2015-10-01'].groupby('id').mean().reset_index()

average_year_price
```

```python
"""
STEP 3: CHANGING COLUMNS NAME AND ADDING COLUMNS WITH SIMILAR DATA.
"""

# Changing Columns Names.
average_year_price = average_year_price.rename(
                                              index=str,
                                               columns= {'price_off_peak_var': 'ave_year_price_offpeak_var', 
                                                'price_peak_var': 'ave_year_price_peak_var',
                                                'price_mid_peak_var': 'ave_year_price_midpeak_var',
                                                'price_off_peak_fix': 'ave_year_price_offpeak_fix',
                                                'price_peak_fix': 'ave_year_price_peak_fix',
                                                'price_mid_peak_fix': 'ave_year_price_midpeak_fix'
                                                })
average_year_price['ave_year_price_offpeak'] = average_year_price['ave_year_price_offpeak_var'] + average_year_price['ave_year_price_offpeak_fix']
average_year_price['ave_year_price_peak'] = average_year_price['ave_year_price_peak_var'] + average_year_price['ave_year_price_peak_fix']
average_year_price['ave_year_price_midpeak'] = average_year_price['ave_year_price_midpeak_var'] + average_year_price['ave_year_price_midpeak_fix']

average_6m_price = average_6m_price.rename(index=str,
                                           columns={'price_off_peak_var': 'ave_6m_price_offpeak_var', 
                                                'price_peak_var': 'ave_6m_price_peak_var',
                                                'price_mid_peak_var': 'ave_6m_price_midpeak_var',
                                                'price_off_peak_fix': 'ave_6m_price_offpeak_fix',
                                                'price_peak_fix': 'ave_6m_price_peak_fix',
                                                'price_mid_peak_fix': 'ave_6m_price_midpeak_fix',
                                                })

average_6m_price['ave_6m_price_offpeak'] = average_6m_price['ave_6m_price_offpeak_var'] + average_6m_price['ave_6m_price_offpeak_fix']
average_6m_price['ave_6m_price_peak'] = average_6m_price['ave_6m_price_peak_var'] + average_6m_price['ave_6m_price_peak_fix']
average_6m_price['ave_6m_price_midpeak'] = average_6m_price['ave_6m_price_midpeak_var'] + average_6m_price['ave_6m_price_midpeak_fix']

average_3m_price = average_3m_price.rename(index= str,
                                           columns={'price_off_peak_var': 'ave_3m_price_offpeak_var', 
                                                'price_peak_var': 'ave_3m_price_peak_var',
                                                'price_mid_peak_var': 'ave_3m_price_midpeak_var',
                                                'price_off_peak_fix': 'ave_3m_price_offpeak_fix',
                                                'price_peak_fix': 'ave_3m_price_peak_fix',
                                                'price_mid_peak_fix': 'ave_3m_price_midpeak_fix',
                                                })
average_3m_price['ave_3m_price_offpeak'] = average_3m_price['ave_3m_price_offpeak_var'] + average_3m_price['ave_3m_price_offpeak_fix']
average_3m_price['ave_3m_price_peak'] = average_3m_price['ave_3m_price_peak_var'] + average_3m_price['ave_3m_price_peak_fix']
average_3m_price['ave_3m_price_midpeak'] = average_3m_price['ave_3m_price_midpeak_var'] + average_3m_price['ave_3m_price_midpeak_fix']
```

```python
"""
STEP 4: MARGING THE CLIENT CHURN WITH THE PRICE DATASET. 
"""
new_price = pd.merge(average_year_price, average_6m_price, on='id')
new_price = new_price.merge(average_3m_price,how='inner', on='id')
price_analysis = pd.merge(new_price , client_df[['id','churn']], how='inner', on='id')
#.drop(columns=['price_off_peak_var', 'price_peak_var', 'price_mid_peak_var', 'price_off_peak_fix', 'price_peak_fix', 'price_mid_peak_fix'],axis=1)

price_analysis
```

```python
"""
STEP 5: CALCULATING THE CORRELATION BETWEEN PRICE AND CHURN.
"""
corr = price_analysis.corr()

corr
```

```python
from numpy.ma.core import size
"""
STEP 6: HEAT MAP DIAGRAM FOR THE PRICE ANALYSIS CORRELATION. 
"""
plt.figure(figsize=(20,18))
sns.heatmap(corr, xticklabels=corr.columns.values, yticklabels=corr.columns.values, annot = True, annot_kws={'size':10})
# Axis ticks size
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

Observation: From the graph above we can observe that there seem to be a positive relationship between churn and price, but this relationship can also be seen to be a weak relationship between churn and price, and to indicate that price is a major factor for churn prediction further feature enginnering is needed.
