## Price Sencitivity Hypothesis Testing
---

In this section we will be defining what price sencitivity, how we are mearsuring it with our dataset, and explaining how changes in price of our services can affect the churning rate of our client. 

### What is Price Sensitivity?

Price sensitivity can be define as the degree of responses of a feature to a change in the price. This can be viewed as measuring the price elasticity of churn as opposed to the useal term of demand and supply. 

In testing our client hypothesis on whether churn is faced by change in price and to test the whether discount can make customers that are likely to churn remain in our company. 

### Measuring rice Sensitivity to Churn Rate.

To measure the price sensitivity of our data we will be calculating the percentage change in price of churn and active customers. I use this method to calculate the price sensitivity of our client customers base off the data set that was give to use, the assumed way to calculate price sensitivity is to use the formular use to calculate price elasticity, which is percentage change in quantity divided by percentage change in price and in other to calculate this we will need the date at which each customer churn of which we do not have in our customer dataset. 

The first step in measuring and comparing the price sensitivity of customer that have churn and active is to marge the price dataset with the client dataset column of churn customer. This is followed by converting the price date from strick to datetime format, this will make it easy sort the data by month. After this, the data will be split into two dataframe i.e churn and active, and then we will average for the 6 different price we have in the price dataset. next, we will make the date as the index for the two dataset, this will enable us to resample the idex from the daily price to an average monthly price index. Lastly, we will be calculating the percentage change in price for both churn and active. 


Evaluating whether our client customers are price sensitivity or not, we will be plotting a graph showing the percentage of change of churn compare with active customers and we can observe whether there is a significant different between churn and active customers. 

```python
"""
STEP 1: MARGING THE PRICE DATASET WITH CLIENT CHURN COLUMN.
"""
# EXAMING THE ID COLUMN OF BOTH DATASET
pri = price_df['id'] == client_df['id'][0]
# price_df.loc[pri]

# MARGING THE TWO DATASET 
price_join = pd.merge(price_df, client_df[['id','churn']], on='id')
price_join.loc[pri]
```

```python
"""
STEP 2: CONVERTING THE STRICK DATASET TO A DATE_TIME FORMATE.
"""

price_join['price_date'] = pd.to_datetime(price_join['price_date'])

price_join.info()
```

```python
"""
STEP 3: SPLITING THE DATASET INTO TWO CHURN AND ACTIVE
"""

# Churn customer
churn_price = price_join[price_join['churn'] == 1]

# Setting the date as the index
churn_price.set_index('price_date', inplace=True)

churn_price
```

```python
# ACTIVE CUSTOMERS

active_price = price_join[price_join['churn'] == 0]

# setting price_date as the index
active_price.set_index('price_date', inplace=True)

active_price
```

```python
"""
STEP 4: CALCULATING THE AVERAGE PRICE
"""

# Churn average price 
churn_price['av_price'] = churn_price.mean(axis=1)

churn_price
```

```python
# active customer average price

active_price['av_price'] = active_price.mean(axis=1)

active_price
```

```python
# The Next step is to get the average price for each month of both churn
# and active customer 

# Churn Customer

churn_monthly_price = pd.DataFrame(churn_price[ 'av_price'])
churn_monthly_price = churn_monthly_price.groupby(by='price_date').mean()
churn_monthly_price
```

```python
# Active Customers
active_monthly_price = pd.DataFrame(active_price[ 'av_price'])
active_monthly_price = active_monthly_price.groupby(by='price_date').mean()
active_monthly_price
```

```python
"""
STEP 5: CALCULATING PERCENTAGE CHANGE IN PRICE
"""
# Churn percent change over the year
churn_pct = churn_monthly_price.pct_change()

churn_pct 
```

```python
# Active customers percentage change in price
active_pct = active_monthly_price.pct_change()
active_pct
```

```python
# plotting the relationship between the churn and active customer.
# This is to compare the distance between the two dataset on monthly price.
plt.plot(active_monthly_price, label='Active Customer', color='green')
plt.plot(churn_monthly_price, label='Churn Customer', color='steelblue')

plt.legend(title="CUSTOMER SEGMENT")

#adding axes labels and title
plt.ylabel('Monthly Price', fontsize=14)
plt.xlabel('Price_Date', fontsize=14)
plt.title('Churn and Active Customer Monthly Price', fontsize=16)

#display plot
plt.show()
```
